```python
import pandas as pd
import json
from tqdm import tqdm

def create_chat_dataset_from_twcs(csv_path: str, output_path: str):
    """
    Parses the TWCS dataset and formats it into a JSONL file with a
    'messages' structure for fine-tuning chat models.

    Args:
        csv_path (str): The path to the input twcs.csv file.
        output_path (str): The path to save the output .jsonl file.
    """
    print(f"Loading data from {csv_path}...")
    try:
        # Load the dataset
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f"Error: The file '{csv_path}' was not found.")
        return

    # --- Pre-processing for efficient lookups ---
    # Drop rows where the reply ID is missing, as they can't form a conversation
    df.dropna(subset=['in_response_to_tweet_id'], inplace=True)

    # Convert tweet ID to integer for consistent matching
    df['in_response_to_tweet_id'] = df['in_response_to_tweet_id'].astype(int)

    # Create a dictionary-like index for fast tweet lookups
    tweets_indexed = df.set_index('tweet_id')

    # --- Identify and pair conversations ---
    conversations = []

    # Filter for agent replies, which are the start of our pairing logic
    agent_replies = df[df['inbound'] == False]

    print("Pairing customer requests with agent replies...")
    # Use tqdm for a user-friendly progress bar
    for _, reply_row in tqdm(agent_replies.iterrows(), total=agent_replies.shape[0]):
        try:
            # Find the original tweet the agent is replying to
            original_tweet_id = reply_row['in_response_to_tweet_id']
            original_tweet = tweets_indexed.loc[original_tweet_id]

            # Ensure the original tweet was from a customer (inbound)
            if original_tweet['inbound']:
                # Format the conversation in the specified "messages" structure
                conversation = {
                    "messages": [
                        {"role": "user", "content": original_tweet['text']},
                        {"role": "assistant", "content": reply_row['text']}
                    ]
                }
                conversations.append(conversation)

        except KeyError:
            # This happens if the original tweet ID is not in our dataset, we can safely ignore it
            continue

    # --- Save the formatted data to a JSONL file ---
    print(f"Saving {len(conversations)} conversations to {output_path}...")
    with open(output_path, 'w', encoding='utf-8') as f:
        for convo in conversations:
            f.write(json.dumps(convo) + '\n')

    print("Processing complete.")
    print(f" Successfully created {output_path}")


# --- Example Usage ---
if __name__ == "__main__":
    # Define the path to your downloaded TWCS dataset
    # You can get this from Kaggle: https://www.kaggle.com/datasets/thoughtvector/customer-support-on-twitter
    input_csv_file = '/content/twcs/twcs.csv'

    # Define the desired name for your output file
    output_jsonl_file = 'customer_support_conversations.jsonl'

    create_chat_dataset_from_twcs(csv_path=input_csv_file, output_path=output_jsonl_file)
```

```python
import pandas as pd
import json
from tqdm import tqdm
from typing import Set, Optional

# A predefined set of major air and travel brands found in the TWCS dataset.
# You can easily add or remove brands from this set.
AIR_TRAVEL_BRANDS = {
    'AmericanAir', 'AirAsiaSupport', 'British_Airways', 'SouthwestAir',
    'KLM', 'Delta', 'JetBlue', 'United', 'Tesco', 'Lufthansa', 'easyJet',
    'flySFO', 'hertz', 'HiltonSuggests', 'Expedia', 'AirbnbHelp',
    'AlaskaAir', 'WestJet'
}

def create_chat_dataset_from_twcs(
    csv_path: str,
    output_path: str,
    filter_authors: Optional[Set[str]] = None
):
    """
    Parses the TWCS dataset and formats it into a JSONL file with a
    'messages' structure, with an option to filter for specific brands.

    Args:
        csv_path (str): The path to the input twcs.csv file.
        output_path (str): The path to save the output .jsonl file.
        filter_authors (Set[str], optional): A set of author_ids (brands)
                                             to include. If None, all brands are processed.
                                             Defaults to None.
    """
    print(f"Loading data from {csv_path}...")
    try:
        df = pd.read_csv(csv_path)
    except FileNotFoundError:
        print(f"Error: The file '{csv_path}' was not found.")
        return

    # --- Pre-processing for efficient lookups ---
    df.dropna(subset=['in_response_to_tweet_id'], inplace=True)
    df['in_response_to_tweet_id'] = df['in_response_to_tweet_id'].astype(int)
    tweets_indexed = df.set_index('tweet_id')

    # --- Identify agent replies and apply the filter ---
    agent_replies = df[df['inbound'] == False]

    if filter_authors:
        print(f"Filtering for conversations with: {', '.join(filter_authors)}")
        initial_reply_count = len(agent_replies)
        agent_replies = agent_replies[agent_replies['author_id'].isin(filter_authors)]
        print(f"Reduced agent replies from {initial_reply_count} to {len(agent_replies)} based on filter.")
    else:
        print("No brand filter applied, processing all conversations.")

    # --- Pair conversations ---
    conversations = []
    print("Pairing customer requests with filtered agent replies...")
    for _, reply_row in tqdm(agent_replies.iterrows(), total=agent_replies.shape[0]):
        try:
            original_tweet_id = reply_row['in_response_to_tweet_id']
            original_tweet = tweets_indexed.loc[original_tweet_id]

            if original_tweet['inbound']:
                conversation = {
                    "messages": [
                        {"role": "user", "content": original_tweet['text']},
                        {"role": "assistant", "content": reply_row['text']}
                    ]
                }
                conversations.append(conversation)
        except KeyError:
            continue

    # --- Save the formatted data to a JSONL file ---
    print(f"Saving {len(conversations)} conversations to {output_path}...")
    with open(output_path, 'w', encoding='utf-8') as f:
        for convo in conversations:
            f.write(json.dumps(convo) + '\n')

    print("Processing complete.")
    print(f" Successfully created {output_path} with {len(conversations)} conversations.")


# --- Example Usage ---
if __name__ == "__main__":
    # Define the path to your downloaded TWCS dataset
    input_csv_file = '/content/twcs/twcs.csv'

    # Define the desired name for your filtered output file
    output_jsonl_file = 'airline_travel_support_conversations.jsonl'

    # Run the function, passing the set of airline/travel brands to the filter
    create_chat_dataset_from_twcs(
        csv_path=input_csv_file,
        output_path=output_jsonl_file,
        filter_authors=AIR_TRAVEL_BRANDS
    )
```
